```
class Solution:
    def getConcatenation(self, nums: list[int]) -> list[int]:
        return nums + nums
```

```
class Solution:
    def shuffle(self, nums: list[int], n: int) -> list[int]:
        ans = []
        for i in range(len(nums) // (len(nums)//n)):
            ans.append(nums[i])
            ans.append(nums[i + n])
        return ans
```

```
class Solution:
    def findMaxConsecutiveOnes(self, nums: list[int]) -> int:
        len = count = 0
        for i in nums:
            if i == 1:
                count += 1
            else:
                count = 0
            len = max(len, count)
        return len
```
